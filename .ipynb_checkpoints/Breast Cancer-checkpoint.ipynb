```python
df.replace({'diagnosis' : {'M' : 1, 'B' : 0}}, inplace=True)
```

```python
df.head()
```

```python
df.groupby(df['diagnosis']).mean()
```

```python
# Creating features and target
X = df.drop(columns=['id', 'diagnosis'])
Y = df['diagnosis']
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Splitting the data
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.30, stratify=Y, random_state=10)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Training the model

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
X_train_predict = model.predict(X_train)
print(X_train_predict)
```

```python
# Checking the accuracy of training prediction
X_train_accuracy = accuracy_score(Y_train, X_train_predict)
print("The Accuracy score on training data is : ",X_train_accuracy)
```

```python
X_test_predict = model.predict(X_test)
print(X_test_predict)
```

```python
X_test_accuracy = accuracy_score(Y_test, X_test_predict)
print("The accuracy score of test data is : ",X_test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test, X_test_predict)
```

```python
pd.DataFrame(confusion_matrix(Y_test, X_test_predict), columns=['Benign Tumour', 'Malignant Tumour'], index=['Actaully Benign', 'Actually Malignant'])
```

```python
def prediction(x):
    input_arr_reshaped = x.reshape(1, -1)
#     print(input_arr_reshaped)
    predict = model.predict(input_arr_reshaped)
    return predict
```
